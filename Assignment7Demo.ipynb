```python
## Import necessary packages:
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
## Load the dataset:
MarketingData = pd.read_csv("/content/Marketing_Data.csv")
```

```python
## Review the first five rows of the dataset:
MarketingData.head(5)
```

```python
## Now, we want to compute linear regression: Y = a+bX taking sales as Y and youtube (ad. cost) as X

## Predictor (independent) variable
X = MarketingData.iloc[:, 0:1]
```

```python
## Response (dependent) variable:
Y = MarketingData.iloc[:, 3:4]
```

```python
## Compute Linear Regression:
LR = LinearRegression()
Reg = LR.fit(X,Y)
```

```python
## Intercept (a)
print(Reg.intercept_)
```

Interpretation: This is the predicted value of Y (sales) when X (spending in youtube) = 0.

```python
## Coefficient or slope (b)
print(Reg.coef_)
```

Interpretation: For each change of one unit in X, Y will change by 0.04820139 units.

To plot the best fitting line and to find R-squared, we want to compute the predicted responses first.

```python
## Predicted values:
Y_pred = LR.predict(X)
```

Now, create a scatterplot and best fitting line.

```python
## Scatterplot:
plt.scatter(X,Y)

## Regression (best fitting) line:
plt.plot(X, Y_pred, color = "red")
plt.show()
```

```python
## Calculate R-squared now:
r2_score(Y, Y_pred)
```

R-squared gives the level of accuracy the regression line fits the data. It measures the "goodness of fit", meaning how well does the regression line approximates the actual data.
